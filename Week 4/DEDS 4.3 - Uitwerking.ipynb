Data in datawarehouse zetten

```python
QUERY = pd.DataFrame(pd.read_sql_query("SELECT * FROM returned_item", go_sales))

all_queries = ""

for index, row in QUERY.iterrows():
    try:
        query = f"""
        INSERT INTO RETURN_DATE(RETURN_DATE_RETURN_DATE_number) 
        VALUES ('{row['RETURN_DATE']}')
        """
        
        all_queries += query
    except Exception as e:
        print(f"Error executing query: {query}")
        print(f"Error details: {e}")

print(all_queries)
```
